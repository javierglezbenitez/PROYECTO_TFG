```python
import pandas as pd
import urllib.parse

df = pd.read_csv(
    r"C:\Users\cgsos\Documents\Cuarto\TFG\FLICKR\datos\LLM\la_gomera_ocio\dirty_dataset\restauracion_la_gomera_sucio.csv",
    sep=";"
)

df = df.rename(columns={
    "direccion_municipio": "municipio",
    "establecimiento_nombre_comercial": "nombre",
    "establecimiento_grupo": "grupo"
})

df["municipio"] = df["municipio"].astype(str).str.strip()
df["nombre"] = df["nombre"].astype(str).str.strip()
df["grupo"] = df["grupo"].astype(str).str.strip()


df = df[df["municipio"] != "_U"]

def maps_url(nombre, municipio):
    q = urllib.parse.quote(f"{nombre} {municipio} La Gomera")
    return f"https://www.google.com/maps/search/{q}"

rows = []

for municipio, sub in df.groupby("municipio"):
    sub_unique = sub.drop_duplicates(subset=["nombre"])
    seleccion = sub_unique.head(2)

    for _, row in seleccion.iterrows():
        rows.append({
            "municipio": municipio,
            "nombre": row["nombre"],
            "grupo": row["grupo"],
            "url_maps": maps_url(row["nombre"], municipio)
        })

restaurantes = pd.DataFrame(rows).sort_values(["municipio", "nombre"])

restaurantes
```

```python
restaurantes.to_csv(r"C:\Users\cgsos\Documents\Cuarto\TFG\FLICKR\datos\LLM\la_gomera_ocio\clean_dataset\restaurantes_la_gomera.csv", index=False, encoding="utf-8")
```

```python
import pandas as pd
import urllib.parse


df = pd.read_csv(
    r"C:\Users\cgsos\Documents\Cuarto\TFG\FLICKR\datos\LLM\la_gomera_ocio\dirty_dataset\alojamientos_la_gomera_sucio.csv",
    sep=",",
    engine="python",
    on_bad_lines="skip"
)

df = df.rename(columns={
    "nom_municipio": "municipio",
    "etiqueta": "nombre",
    "nom_tipo": "tipo",
    "nom_categor": "categoria"
})


df["municipio"] = df["municipio"].astype(str).str.strip()
df["nombre"] = df["nombre"].astype(str).str.strip()
df["tipo"] = df["tipo"].astype(str).str.strip()
df["categoria"] = df["categoria"].astype(str).str.strip()

def maps_url(nombre, municipio):
    q = urllib.parse.quote(f"{nombre} {municipio} La Gomera")
    return f"https://www.google.com/maps/search/{q}"

rows = []

for municipio, sub in df.groupby("municipio"):
    sub = sub.drop_duplicates(subset=["nombre"])

    seleccion = sub.head(2)

    for _, row in seleccion.iterrows():
        rows.append({
            "municipio": municipio,
            "nombre": row["nombre"],
            "tipo": row["tipo"],
            "url_maps": maps_url(row["nombre"], municipio)
        })

alojamientos_limpio = pd.DataFrame(rows).sort_values(["municipio", "nombre"])


alojamientos_limpio
```
