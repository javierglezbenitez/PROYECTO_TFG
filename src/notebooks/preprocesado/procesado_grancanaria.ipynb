**RESTAURANTES**

url = https://datos.canarias.es/catalogos/general/dataset/establecimientos-de-restauracion-inscritos-en-el-registro-general-turistico-de-canarias

```python
import pandas as pd
import urllib.parse

df = pd.read_csv(r"C:\Users\cgsos\Documents\Cuarto\TFG\FLICKR\datos\LLM\grancanaria_ocio\dirty_dataset\restaurantes_gran_canaria_sucio.csv")

df["municipio"] = df["municipio"].astype(str).str.strip()
df["nombre"] = df["nombre"].astype(str).str.strip()
df["grupo"] = df["grupo"].astype(str).str.strip()

df = df[df["municipio"] != "_U"]

def maps_url(nombre, municipio):
    q = urllib.parse.quote(f"{nombre} {municipio} Gran Canaria")
    return f"https://www.google.com/maps/search/{q}"

rows = []

for municipio, sub in df.groupby("municipio"):
    sub_unique = sub.drop_duplicates(subset=["nombre"])
    seleccion = sub_unique.head(2)

    for _, row in seleccion.iterrows():
        rows.append({
            "municipio": municipio,
            "nombre": row["nombre"],
            "grupo": row["grupo"],
            "url_maps": maps_url(row["nombre"], municipio)
        })

restaurantes = pd.DataFrame(rows)
restaurantes = restaurantes.sort_values(["municipio", "nombre"])

restaurantes
```

```python
restaurantes.to_csv(r"C:\Users\cgsos\Documents\Cuarto\TFG\FLICKR\datos\LLM\grancanaria_ocio\clean_dataset\restaurantes_gran_canaria.csv", index=False, encoding="utf-8")
```

```python
import pandas as pd
import urllib.parse

df = pd.read_csv(
    r"C:\Users\cgsos\Documents\Cuarto\TFG\FLICKR\datos\LLM\grancanaria_ocio\dirty_dataset\alojamientos_gran_canaria_sucio.csv"
)

df["nom_municipio"] = df["nom_municipio"].astype(str).str.strip()
df["etiqueta"] = df["etiqueta"].astype(str).str.strip()

def maps_url(nombre, municipio):
    q = urllib.parse.quote(f"{nombre} {municipio} Gran Canaria")
    return f"https://www.google.com/maps/search/{q}"

rows = []

for municipio, sub in df.groupby("nom_municipio"):
    sub_unique = sub.drop_duplicates(subset=["etiqueta"])

    seleccion = sub_unique.head(2)

    for _, row in seleccion.iterrows():
        rows.append({
            "municipio": municipio,
            "nombre": row["etiqueta"],
            "tipo": row["nom_tipo"] if pd.notna(row["nom_tipo"]) else "Alojamiento",
            "url_maps": maps_url(row["etiqueta"], municipio)
        })

alojamientos = pd.DataFrame(rows)
alojamientos = alojamientos.sort_values(["municipio", "nombre"])

alojamientos
```
