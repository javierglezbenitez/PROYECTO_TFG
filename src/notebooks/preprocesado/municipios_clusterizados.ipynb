```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
CSV_NIVELES = r"C:\Users\cgsos\Documents\Cuarto\TFG\FLICKR\datos\LLM\municipios_etiquetados_niveles.csv"
df = pd.read_csv(CSV_NIVELES)

df_gc = df[df["isla"].str.lower() == "gran canaria"].copy()

df_gc = df_gc.sort_values("nivel_turistico_municipio")

plt.figure(figsize=(8, 6))

plt.barh(
    df_gc["municipio"],
    df_gc["nivel_turistico_municipio"],
    color=plt.cm.RdYlBu_r(df_gc["nivel_turistico_municipio"] / 4)
)

plt.xlabel("Nivel turístico municipal (1–4)")
plt.ylabel("Municipio")
plt.title("Clasificación turística de municipios de Gran Canaria\n(agregación de clusters)")

plt.xlim(0.5, 4.5)
plt.xticks([1, 2, 3, 4])

plt.tight_layout()
plt.show()
```
